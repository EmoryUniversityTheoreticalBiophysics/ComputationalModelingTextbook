## Problem 2

Suppose you wrote a program for simulating the dynamics of the number of viral particles (virions) of HIV in the body of a human following application of some antiviral drug. These particles (let's denote their number as $v$) are killed by the body at the rate $-r_vv$. At the same time, these particles are released into the blood stream by infected T-cells, the number of which we will denote as $c$, at a rate of $kc$. In their turn, the infected cells are killed by the body's immune system at the rate $-r_cc$. Ordinarily, the virions in blood would infect other cells in the body, but this doesn't happen after the application of the drug. Notice that this is a system of two coupled ordinary differential equations (one for cells, and one for virions), which are very similar to various bacterial growth equations we studied in class. The input to the program consists of the initial conditions (initial number of virions and infected T-cells), the time interval over which the problem should be analyzed, and the value of the rate constants. The program outputs the number of virions and infected T-cells as a function of time. 

First, write down the set of differential equations that describes your model (8 points).  Then List some of the specific steps you would do to verify your program. In particular, think about how the system would behave when $r_c\gg r_v$ or $r_v \gg r_c$ (these are two independent verification steps) -- but you may choose different verification steps as well. (12 points -- 3 points each for each correct, nontrivial verifcation step.)

Note that this problem admits an analytical solution -- feel free to use comparison to the analytical solution as one of your verification steps if you can derive it (but do not attempts to derive it if you do not know what you are doing -- you will waste time).
***

**The mathematical model.** <font color='blue'>The set of differential equations describing the model:
$$ \frac{dv}{dt}= kc - r_v v,\\ \frac{dc}{dt}=-r_c c.$$
This allows now to devise a few verification steps.</font>

**Verification 1**. 
<font color='blue'>Let's first see if the system has any steady states, and what they are. For this, we set all derivatives to zero.
$$0=kc - r_vv\\0=-r_cc$$
This results in only one solution: $c=0$, $v=0$, which is, incidentally, stable. The verification is then to simulate the system from an arbitrary initial condition and with arbitrary parameters, and both $c$ and $v$ must approach the steady state.</font>


**Verification 2.** <font color='blue'>The solution of the second equation above (for the T-cells) is $c(t)=c_0e^{-r_ct}$. Thus the curve for $c$ should be linear on a semi-long plot with the slope of $-r_c$.</font>

**Verification 3**. <font color='blue'>If $r_c\gg r_v$, then cells will die out quickly, resulting in the first equation changing to $\frac{dv}{dt}\approx -r_v v$, which gives a solution $v(t)= v_0 e^{-r_vt}$. Thus the curve for $v$ should be linear on a semi-long plot with the slope of $-r_v$.</font>

**Verification 4**. <font color='blue'>If $r_v\gg r_c$, then cells are decaying slowly, while dynamics of the virions is fast. That is, and at any point in time, the virion number should be in a quasi steady state defined by the current concentration of the cells and by the virion death rate, so that the current production is completely balanced by the current degradation. That is, $kc=r_vv$. Plugging in the expression for $c(t)$, gives $v(t)= \frac{k}{r_v} c_0e^{-r_ct}$. Thus the curve for $v$ should be linear on a semi-long plot with the slope of $-r_c$, and the intercept should be given by the prefactor in the expression.</font>

**Verification 5**. <font color='blue'>(I am providing an extra verification for you, just to show that there are many.) If the virus starts at $v(0)=0$, then the virst equation results in $\frac{dv}{dt}= kc$, so that, in the very beginning of the dynamics, the virus will be increasing with the slope $kc_0$, before it starts decreasing.</font>

## Problem 3

Write the Python code that would solve for the dynamics of the virions in Problem 2. Allow for arbitrary initial conditions. Use the Python built-in ordinary differential equations solver `scipy.integrate.odeint()` for this task. Plot the output $x(t)$. Does it match your expectations? Verify your model using the  verification steps you developed in Problem 2.

Insert cells after the text of this problem and write your code there; make sure it executes. Total 25 point.
***

<font color='blue'>All of the subsequent code cells would need to be written by you; they were not provided in the exam notebook.</font>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

SimulationTime = 20.0
dt = 0.01
t = np.arange(0.0, SimulationTime+dt, dt)

# r.h.s. of the system of ODEs defined above
def HIV_dynamics(Population, t, k, rc, rv):
    dc = - rc*Population[0] 
    dv = k*Population[0] - rv*Population[1] 
    
    return np.array((dc,dv))

# helper function for plotting similar plots
def PlotDynamics(t, Population):
    plt.plot(t, Population[:,0], label='T cells')
    plt.plot(t, Population[:,1], label='Virions')
    plt.xlabel('time')
    plt.ylabel('population')
    plt.legend()
    plt.show()
    
# helper function for plotting similar plots, in semilogy
def PlotDynamicsLog(t, Population):
    plt.semilogy(t, Population[:,0], label='T cells')
    plt.semilogy(t, Population[:,1], label='Virions')
    plt.xlabel('time')
    plt.ylabel('population')
    plt.legend()
    plt.show()
```

```python
# doing the first simulation run at some parameter values
k = 1.0
rc = 1.0
rv = 1.0
Pop0 = np.array((10.0,10.0))
Population = odeint(HIV_dynamics, Pop0, t, args=(k,rc,rv))
PlotDynamics(t,Population)
# notice that the dynamics settles at 0 for both T cells and virions, completing Verification 1
```

```python
# Verification 2
PlotDynamicsLog(t,Population)
# Notice that the T cells population falls linearly on the semilog plot, with the slope of rc=-1, as predicted.
# 
# Recall that exp(3) is about 20. So, with rc=-1, we expect the population to fall by a factor of 20 for t=3. 
# This makes it easier to read the plots and estimate the rate of decay from the plot.
```

```python
# Verification 3
rc = 5.0
rv = 1.0
SimulationTime = 4.0
t = np.arange(0.0, SimulationTime+dt, dt)

Population = odeint(HIV_dynamics, Pop0, t, args=(k,rc,rv))
PlotDynamicsLog(t,Population)
# Notice that the virion population falls linearly on the semilog plot, with the slope of rv=-1, as predicted. 
```

```python
# Verification 4
rc = 1
rv = 5
SimulationTime = 4.0
t = np.arange(0.0, SimulationTime+dt, dt)

Population = odeint(HIV_dynamics, Pop0, t, args=(k,rc,rv))
PlotDynamicsLog(t,Population)
# Notice that virions population falls linearly on the semilog plot, with the slope of rc=-1, as predicted.
```

```python
# Verification 5
rc = .1
rv = 1
SimulationTime = 5.0
t = np.arange(0.0, SimulationTime+dt, dt)

Pop0 = np.array((10.0,0.0))
Population = odeint(HIV_dynamics, Pop0, t, args=(k,rc,rv))
PlotDynamics(t,Population)
# Notice that virions population increases linearly (before decreasing) with the slope about k*c(0)=10
```
